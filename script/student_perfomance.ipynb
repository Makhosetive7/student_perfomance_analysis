```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
dataFrame['Gender'] = dataFrame['Gender'].map({0:'Female', 1:'Male'})
dataFrame['Tutoring'] = dataFrame['Tutoring'].map({0:'No', 1:'Yes'})
dataFrame['ParentalSupport'] = dataFrame['ParentalSupport'].map({0:'Low', 1:'Medium', 2:'High'})
```

```python
# total number of students
total_students = dataFrame.shape[0]
print(f'Total number of students: {total_students}')
```

```python
# student grade level distribution
plt.figure(figsize=(10,6))
sns.countplot(x='GradeClass', data=dataFrame, palette='viridis')
plt.title('Student Grade Level Distribution')
plt.xlabel('Grade Level')
plt.ylabel('Number of Students')
plt.show()
```

```python
# Gender distribution
plt.figure(figsize=(8,5))
sns.countplot(x='Gender', data=dataFrame, palette='pastel')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

### What we are looking for

- Percentage of students participating in each activity

- How does GPA differ between students in sports vs non-sports?

- How does GPA differ between students in music vs non-music?

- How does GPA differ for students who volunteer vs those who dont?

- Do students involved in multiple activities have higher GPAs?

- Are boys or girls more active in extracurriculars?

- Which grade levels participate most in extracurriculars?

- Do students in extracurriculars have fewer absences on average?


```python
# Percentage of students participating in each activity
activities = ['Sports', 'Music', 'Volunteering']
percentages = dataFrame[activities].mean() * 100
print(percentages)

# Plot
percentages.plot(kind='bar', color='skyblue', title='Percentage of Students in Extracurriculars')
plt.xlabel('Activity')
plt.ylabel('Percentage (%)')
plt.show()
```

```python
# What percentage participate in Extracurricular activities?
Music_participation = dataFrame['Music'].value_counts(normalize=True) * 100
Sports_participation = dataFrame['Sports'].value_counts(normalize=True) * 100
Voluntering_participartion = dataFrame['Volunteering'].value_counts(normalize=True)* 100
print(Music_participation)
print(Sports_participation)
print(Voluntering_participartion)
```

```python
# How does GPA differ between students in sports vs non-sports?
plt.figure(figsize=(20,6))
sns.boxplot(x='Sports', y='GPA', data=dataFrame, palette='Set2')
plt.title('GPA Distribution: Sports vs Non-Sports Participants')
plt.xlabel('Participates in Sports (0=No, 1=Yes)')
plt.ylabel('GPA')
plt.show()  
```

```python
# How does GPA differ between students in music vs non-music?
plt.figure(figsize=(20, 6))
sns.boxplot(x='Music', y='GPA', data=dataFrame, palette='Set3')
plt.title('GPA Distribution: Music vs Non-Music Participants')
plt.xlabel('Participates in Music (0=No, 1=Yes)')
plt.ylabel('GPA')
plt.show()
```

```python
# How does GPA differ for students who volunteer vs those who dont?
plt.figure(figsize=(20, 6))
sns.boxplot(x='Volunteering', y='GPA', data=dataFrame, palette='Set1')
plt.title('GPA Distribution: Volunteering vs Non-Volunteering Participants')
plt.xlabel('Participates in Volunteering (0=No, 1=Yes)')
plt.ylabel('GPA')
plt.show()  
```

```python
# Do students involved in multiple activities have higher GPAs?
dataFrame['TotalActivities'] = dataFrame[activities].sum(axis=1)
plt.figure(figsize=(10,6))
sns.boxplot(x='TotalActivities', y='GPA', data=dataFrame, palette='coolwarm')
plt.title('GPA vs Number of Extracurricular Activities')
plt.xlabel('Number of Activities')
plt.ylabel('GPA')
plt.show()
```

```python
# Are boys or girls more active in extracurriculars?
plt.figure(figsize=(10,6))
sns.countplot(x='Sports', hue='Gender', data=dataFrame, palette='Set1')
plt.title('Sports Participation by Gender')
plt.xlabel('Sports Participation (0=No, 1=Yes)')
plt.ylabel('Number of Students')
plt.show()
```

```python
# Which grade levels participate most in extracurriculars?
fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharey=True)

# Music Participation
sns.countplot(x='GradeClass', hue='Music', data=dataFrame, palette='Set2', ax=axes[0])
axes[0].set_title('Music Participation by Grade Level')
axes[0].set_xlabel('Grade Level')
axes[0].set_ylabel('Number of Students')

# Sports Participation
sns.countplot(x='GradeClass', hue='Sports', data=dataFrame, palette='Set2', ax=axes[1])
axes[1].set_title('Sports Participation by Grade Level')
axes[1].set_xlabel('Grade Level')

# Volunteering Participation
sns.countplot(x='GradeClass', hue='Volunteering', data=dataFrame, palette='Set2', ax=axes[2])
axes[2].set_title('Volunteering Participation by Grade Level')
axes[2].set_xlabel('Grade Level')

plt.tight_layout()
plt.show()
```

```python
# Do students in extracurriculars have fewer absences on average?

plt.figure(figsize=(18, 5))

# Music
plt.subplot(1, 3, 1)
sns.boxplot(x='Music', y='Absences', data=dataFrame, palette='Set2')
plt.title('Absences by Music paticipants')
plt.xlabel('Music paticipants')

# Sports
plt.subplot(1, 3, 2)
sns.boxplot(x='Sports', y='Absences', data=dataFrame, palette='Set2')
plt.title('Absences by Sports paticipants')
plt.xlabel('Sports paticipants')

# Volunteering
plt.subplot(1, 3, 3)
sns.boxplot(x='Volunteering', y='Absences', data=dataFrame, palette='Set2')
plt.title('Absences by Volunteering paticipants')
plt.xlabel('Volunteering paticipants')

plt.tight_layout()
plt.show()
```

### Tutoring & Parental Support (10 Questions)

- Do students receiving tutoring have higher GPAs?

- Do students with parental support have higher GPAs?

- How does tutoring affect students with low study time?

- Does tutoring reduce absences?

- How does parental education influence parental support?

- Are there gender differences in receiving tutoring?

- Are there ethnicity differences in parental support?

- Does tutoring impact GPA differently across grade levels?

- How does high parental support correlate with GPA improvement?

- Do students with tutoring show better study time habits?

```python
# Do students receiving tutoring have higher GPAs?
plt.figure(figsize=(20,6))
sns.boxplot(x='Tutoring', y='GPA', data=dataFrame, palette='Set3')
plt.title('GPA Distribution: Tutored vs Non-Tutored Students')
plt.xlabel('Receiving Tutoring (No=0, Yes=1)')
plt.ylabel('GPA')
plt.show()
```

```python
# Do students with parental support have higher GPAs?
plt.figure(figsize=(20,6))
sns.boxplot(x='ParentalSupport', y='GPA', data=dataFrame, palette='Set3')
plt.title('GPA Distribution by Level of Parental Support')
plt.xlabel('Parental Support Level (Low=0, Medium=1, High=2')
plt.ylabel('GPA')
plt.show()
```

```python
# How does tutoring affect students with low study time?
low_study = dataFrame[dataFrame['StudyTimeWeekly'] == 0]
plt.figure(figsize=(20,6))
sns.boxplot(x='Tutoring', y='GPA', data=low_study, palette='Set1')
plt.title('GPA for Low Study Time Students: Tutored vs Non-Tutored')
plt.xlabel('Receiving Tutoring (No=0, Yes=1)')
plt.ylabel('GPA')   
plt.show()
```

```python
# Does tutoring reduce absences?
plt.figure(figsize=(20,6))
sns.boxplot(x='Tutoring', y='Absences', data=dataFrame, palette='Set2')
plt.title('Absences: Tutored vs Non-Tutored Students')
plt.xlabel('Receiving Tutoring (No=0, Yes=1)')
plt.ylabel('Number of Absences')
plt.show()
```

```python
# How does parental education influence parental support?
plt.figure(figsize=(20,6))
sns.countplot(x='ParentalSupport', hue='ParentalEducation', data=dataFrame, palette='Set2')
plt.title('Parental Support by Parental Education Level')
plt.xlabel('Parental Support Level')    
plt.ylabel('Number of Students')
plt.show()
```

```python
# Are there gender differences in receiving tutoring?
sns.catplot(x='Tutoring', hue='Gender', col='GradeClass', data=dataFrame, kind='count', palette='Set1', height=4, aspect=0.8)
plt.subplots_adjust(top=0.85)
plt.suptitle('Tutoring Participation by Gender Across Grades')
plt.show()
```

```python
# Are there ethnicity differences in parental support?
sns.catplot(x='ParentalSupport', hue='Ethnicity', col='GradeClass', data=dataFrame, kind='count', palette='Set2', height=4, aspect=0.8)
plt.subplots_adjust(top=0.85)
plt.suptitle('Parental Support by Ethnicity Across Grades')
plt.show() 
```

```python
#  Does tutoring impact GPA differently across grade levels?
plt.figure(figsize=(20,6))
sns.boxplot(x='GradeClass', y='GPA', hue='Tutoring', data=dataFrame, palette='Set3')
plt.title('GPA by Grade Level and Tutoring Status')
plt.xlabel('Grade Level')
plt.ylabel('GPA')
plt.show()
```

```python
# How does high parental support correlate with GPA improvement? 

plt.figure(figsize=(20,3))
sns.barplot(x='ParentalSupport', y='GPA', data=dataFrame, palette='Set1', ci=None)
plt.title('Average GPA by Parental Support Level')
plt.xlabel('Parental Support Level')
plt.ylabel('Average GPA')
plt.show()

```

```python
# Do students with tutoring show better study time habits
plt.figure(figsize=(10,6))
sns.boxplot(x='Tutoring', y='StudyTimeWeekly', data=dataFrame)
plt.title('Study Time Comparison: Tutoring vs No Tutoring')
plt.xlabel('Tutoring (0=No, 1=Yes)')
plt.ylabel('Study Time per Week (hours)')
plt.show()
```

### Combined Effects & Contributions (10 Questions)

- Do students with both tutoring and parental support perform best?

- Do students with tutoring and extracurricular activities have higher GPAs?

- Do students with parental support and extracurricular involvement perform better?

- Does combining tutoring, parental support, and extracurriculars maximize GPA?

- Which combination has the largest effect on reducing absences?

- Does parental support boost the benefits of extracurricular participation?

- How does study time contribute when tutoring and activities are combined?

- Are students with no support or no extracurricular involvement at a GPA disadvantage?

```python
#  Do students with both tutoring and parental support perform best?
plt.figure(figsize=(20,6))
sns.boxplot(x='ParentalSupport', y='GPA', hue='Tutoring', data=dataFrame, palette='Set2')
plt.title('GPA by Parental Support Level and Tutoring Status')
plt.xlabel('Parental Support Level')
plt.ylabel('GPA')
plt.show()
```

```python
plt.figure(figsize=(24, 6))

# Plot 1: Parental Support + Sports
plt.subplot(1, 3, 1)
sns.boxplot(x='ParentalSupport', y='GPA', hue='Sports', data=dataFrame, palette='Set2')
plt.title('GPA by Parental Support & Sports')
plt.xlabel('Parental Support Level')
plt.ylabel('GPA')
plt.legend(title='Sports', loc='upper right')

# Plot 2: Sports + Tutoring
plt.subplot(1, 3, 2)
sns.boxplot(x='Sports', y='GPA', hue='Tutoring', data=dataFrame, palette='Set3')
plt.title('GPA by Sports & Tutoring')
plt.xlabel('Sports Participation (0=No, 1=Yes)')
plt.ylabel('GPA')
plt.legend(title='Tutoring', loc='upper right')

# Plot 3: Parental Support + Tutoring
plt.subplot(1, 3, 3)
sns.boxplot(x='ParentalSupport', y='GPA', hue='Tutoring', data=dataFrame, palette='Set2')
plt.title('GPA by Parental Support & Tutoring')
plt.xlabel('Parental Support Level')
plt.ylabel('GPA')
plt.legend(title='Tutoring', loc='upper right')

plt.tight_layout()
plt.show()
```

```python
# Does combining tutoring, parental support, and extracurriculars maximize GPA?
plt.figure(figsize=(20,6))
dataFrame['Extracurriculars'] = dataFrame[activities].sum(axis=1)
sns.boxplot(x='Extracurriculars', y='GPA', hue='Tutoring', data=dataFrame, palette='Set3')
plt.title('GPA by Number of Extracurriculars and Tutoring Status')
plt.xlabel('Number of Extracurricular Activities')
plt.ylabel('GPA')
plt.show()
```

```python
# Does parental support boost the benefits of extracurricular participation?
plt.figure(figsize=(20,6))
sns.boxplot(x='ParentalSupport', y='GPA', hue='Extracurriculars', data=dataFrame, palette='Set2')
plt.title('GPA by Parental Support Level and Extracurricular Participation')
plt.xlabel('Parental Support Level')
plt.ylabel('GPA')
plt.legend(title='Number of Extracurriculars', loc='upper right')
plt.show()
```

```python
# How does study time contribute when tutoring and activities are combined?
plt.figure(figsize=(20,10))
sns.scatterplot(x='StudyTimeWeekly', y='GPA', hue='TotalActivities', style='Tutoring', data=dataFrame, palette='Set1', s=100)
plt.title('GPA vs Study Time: Impact of Activities and Tutoring')
plt.xlabel('Study Time per Week (hours)')
plt.ylabel('GPA')
plt.legend(title='Total Activities & Tutoring', loc='upper left')       
plt.show()
```

```python
# Are students with no support or no extracurricular involvement at a GPA disadvantage?
plt.figure(figsize=(20,6))
sns.boxplot(x='TotalActivities', y='GPA', hue='ParentalSupport', data=dataFrame, palette='Set3')
plt.title('GPA by Number of Extracurricular Activities and Parental Support')
plt.xlabel('Number of Activities')
plt.ylabel('GPA')
plt.legend(title='Parental Support Level', loc='upper right')
plt.show()
```
